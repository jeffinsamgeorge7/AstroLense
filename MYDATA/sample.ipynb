```python
import matplotlib.pyplot as plt
from astropy.io import fits

# Specify the FITS file path
fits_file = "HorseHead.fits"

# Open the FITS file
with fits.open(fits_file) as hdul:
  # Get the image data
  image_data = hdul[0].data

# Display the image using matplotlib
plt.imshow(image_data, cmap="gray")  # Adjust 'cmap' for color images
plt.colorbar()


plt.show()
```

```python
from astropy.io import fits
import matplotlib.pyplot as plt

# Specify the FITS file path
fits_file = "HorseHead.fits"

# Open the FITS file
with fits.open(fits_file) as hdul:
  # Get the image data
  data = hdul[0].data

# Display the image using Matplotlib
plt.imshow(data, cmap="gray")  # Adjust 'cmap' for colormaps

# Optional: Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title(f"Image from {fits_file}")

# Show the plot
plt.show()
```
